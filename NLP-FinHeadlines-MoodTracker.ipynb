The dataset provides a valuable resource for sentiment analysis in the domain of finance and can be used to gain insights into investor sentiment based on news headlines. Further details and in-depth analysis of the dataset can be found in the original research paper mentioned above. The paper offers insights into the methodology used for sentiment annotation and provides additional context for understanding the dataset.

For access to the dataset and any specific usage terms or restrictions, please refer to the [original source](https://www.kaggle.com/datasets/ankurzing/sentiment-analysis-for-financial-news).

### Text cleaning
As with any text preprocessing, we will have to remove punctuation,numbers, special characters and extra whitespaces from our 'Headline' column.  
To achieve this, we can use the below clean_text function.

```python
def clean_text(text):
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove special characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply clean_text to the 'Headline' column and create a new column for the cleaned headlines
findata['Cleaned_Headline'] = findata['Headline'].apply(clean_text)
```

Using some percentage calculations to check how much did the data change.

```python
# Count the number of records where the headline has changed
changed_records = (findata['Headline'] != findata['Cleaned_Headline']).sum()

# Calculate the percentage of changed records
percentage_changed = (changed_records / len(findata)) * 100

print(f"Percentage of changed records: {percentage_changed:.2f}%")
```

### Tokenization
The next step in preprocessing is tokenization. For this step, the nltk library is used and more specifically NLTK's word_tokenize function.  
The headlines are split into individual words

```python
findata['Tokenized_Headline'] = findata['Cleaned_Headline'].apply(word_tokenize)
```

```python
findata['Tokenized_Headline']
```

### Stopwords removal
To further preprocess the text data, we perform stopwords removal.Stopwords are commonly used words in a language that carry little semantic meaning and can be safely removed from the text.  
In this project, we utilize NLTK's stopwords library to obtain a set of English stopwords.
Additionally, we convert all words in the 'Headline' column to lowercase to ensure consistency and avoid duplication caused by case sensitivity. That will also reduce the noise of our later model.

```python
stopwords = set(stopwords.words('english'))
findata['Filtered_Headline'] = findata['Tokenized_Headline'].apply(lambda x: [word for word in x if word.lower() not in stopwords])
```

### Lemmatization
In addition to removing stopwords and converting words to lowercase, another important step in text preprocessing is lemmatization.  
Lemmatization is the process of reducing words to their base or dictionary form. In our project, we use NLTK's WordNetLemmatizer to lemmatize the tokenized headlines. By doing so, we can improve the accuracy of sentiment analysis and avoid duplication caused by variations of the same word.

```python
lemmatizer = WordNetLemmatizer()
findata['Lemmatized_Headline'] = findata['Filtered_Headline'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
```

### Expand contrations
We need to expand contractions to further standartise our text. Contractions are shortened forms of words or phrases that combine two words by replacing one or more letters with an apostrophe.For example, don't is a contraction of do not.  
Contractions are used in headlines to make the language used more concise and conversational.  
The **expand_contractions** function below is used to get to the expanded versions of the mentioned words.

```python
def expand_contractions(text_list):
    expanded_texts = []
    for text in text_list:
        expanded_text = contractions.fix(text)
        expanded_texts.append(expanded_text)
    return expanded_texts

findata['Expanded_Headline'] = findata.apply(lambda row: expand_contractions(row['Lemmatized_Headline']), axis=1)
```

### Further tokenisation and padding
We need to perform additional preprocessing steps on the tokenized headlines using the Tokenizer class from Keras and the pad_sequences function from TensorFlow.

After creating an instance of the Tokenizer class and fitting it on the expanded headlines, the **texts_to_sequences** method is called on the tokenizer object to convert the tokenized headlines into sequences of integers.

The **word_index** attribute of the tokenizer object is accessed to obtain the word-to-index mapping, which represents the vocabulary of the headlines.

Next, the sequences of integers are padded using the **pad_sequences** function from TensorFlow. Padding ensures that all sequences have the same length by adding zeros or truncating them if necessary. This step is important for creating fixed-length input sequences required by many machine learning models.

The padded sequences are stored in the variable X, which will serve as the input data for the sentiment analysis model.

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(findata['Expanded_Headline'])
sequences = tokenizer.texts_to_sequences(findata['Expanded_Headline'])
word_index = tokenizer.word_index
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(sequences)
```

### Label encoding
The encoding process involves assigning a unique numerical value to each sentiment class.  
Label encoding assigns a unique integer value to each sentiment class (e.g., 0 for negative, 1 for neutral, 2 for positive). One-hot encoding, on the other hand, represents each sentiment class as a binary vector with a length equal to the number of classes.
We will use both processes, one-hot encoding is needed for our neural network further down the line.  
For label encoding we will use **LabelEncoder** from the scikit learn library and then to one-hot encode we will use **to_categorical** from Keras.
This will be our target variable in the model.

```python
label_encoder = LabelEncoder()
y = to_categorical(label_encoder.fit_transform(findata['Sentiment']))
```

## Class imbalance handling
To address the issue of class imbalance mentioned above, various techniques can be applied, such as:

- Oversampling: Increasing the number of instances in the minority classes by replicating or generating synthetic samples.
- Undersampling: Reducing the number of instances in the majority class by randomly removing samples.
- Class weighting: Assigning higher weights to the minority classes during model training to give them more importance.

### Oversampling
The technique we are going to use is **SMOTE(Synthetic Minority Over-sampling Technique)** from the imblearn library.

```python
smote = SMOTE()
X_train_oversampled, y_train_oversampled = smote.fit_resample(X_train, y_train)
```

### Undersampling
The technique we are going to use is the **RandomUndersampler** from the imblearn library.

```python
undersampler = RandomUnderSampler()
X_train_undersampled, y_train_undersampled = undersampler.fit_resample(X_train, y_train)
```

By applying over or undersampling techniques only on the training set, we ensure that the model learns from a balanced representation of the classes without compromising the integrity of the testing set and the model's ability to generalize to new, unseen data.

### Class weighting
Here we are going to use sklearn's **class_weight**.

```python
class_weights = class_weight.compute_sample_weight('balanced', y_train)
class_weights_dict = {i: weight for i, weight in enumerate(class_weights)}
```

We will determine later on when we compile the model which technique gives us better accuracy and precision.

## Word vectorization
Word vectorization is a process of converting words or text into numerical vectors, which can be used as input for machine learning algorithms.  
In this project, we utilize a pre-trained Word2Vec model, specifically the "GoogleNews-vectors-negative300.bin" model.  
You can download it to your local machine through this [link](https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit?resourcekey=0-wjGZdNAUop6WykTtMip30g)


The pre-trained Word2Vec model contains word embeddings, which are vector representations of words in a high-dimensional space. These word embeddings capture semantic relationships between words, allowing us to perform mathematical operations on the vectors.

To vectorize the expanded headlines, the code applies the Word2Vec model to each word in the headline and retrieves the corresponding word vector. These word vectors are then averaged to obtain a fixed-length vector representation for each headline. By averaging the word vectors, we obtain a condensed representation of the headline that captures the overall meaning and context of the words used.

The word vectorization step is crucial as it transforms the textual data into a numerical format that can be processed by machine learning models. This allows us to apply various classification algorithms to predict the sentiment of the financial news headlines based on the vectorized representations.  
We are using **KeyedVectors** from the gensim library to load the model

```python
model_path = '/Volumes/HD-SL1 1/GoogleNews-vectors-negative300.bin'
model = KeyedVectors.load_word2vec_format(model_path, binary=True)
```

Next, we need an embedding matrix.An embedding matrix is a numerical representation of words or tokens in a text dataset. It captures the relationships between words by assigning each word a vector representation. The embedding matrix is initialized with random values or pre-trained word embeddings, allowing machine learning models to understand the meanings and relationships of words in the dataset.  
We are going to initialise ours with the pre-trained Word2Vec model.

```python
embedding_matrix = np.zeros((len(word_index) + 1, 300))
for word, i in word_index.items():
    if word in model:
        embedding_matrix[i] = model[word]
```

## Model architecture
The architecture of the model has undergone multiple iterations, and this current version represents the configuration that I find most suitable.  
Let's explain it.  
The choice of these layers is based on their effectiveness in handling sequential data and their ability to capture complex patterns and relationships. The embedding layer utilizes pre-trained word embeddings to leverage semantic information. The convolutional layer extracts local features, while the max pooling layer reduces dimensionality. The bidirectional LSTM layer captures long-term dependencies, and the dropout layer prevents overfitting. Finally, the dense layer provides the output classification probabilities.

```python
model = Sequential()
model.add(Embedding(len(word_index) + 1, 300, weights=[embedding_matrix], input_length=X.shape[1], trainable=False))
model.add(Conv1D(128, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=4))
model.add(Bidirectional(LSTM(256)))
model.add(Dropout(0.3))
model.add(Dense(3, activation='softmax'))
```

## Model training and evaluation

```python
optimizers = [RMSprop(), Adagrad(), Adadelta(), Adam()]

results = []  # List to store the results for each optimizer

# Iterate over each optimizer and train the model
for optimizer in optimizers:
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
    history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=64)
    # Evaluate the model and store the results
    loss, accuracy, precision, recall = model.evaluate(X_test, y_test)
    f1_score = 2 * (precision * recall) / (precision + recall)
    results.append({
        'Optimizer': optimizer.__class__.__name__,
        'Test Loss': loss,
        'Test Accuracy': accuracy,
        'Test Precision': precision,
        'Test Recall': recall,
        'Test F1 Score': f1_score
    })

# Create a DataFrame from the results
df_imbalanced = pd.DataFrame(results)
```

```python
results = []  # List to store the results for each optimizer

# Iterate over each optimizer and train the model
for optimizer in optimizers:
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
    history = model.fit(X_train_oversampled, y_train_oversampled, epochs=10, validation_data=(X_test, y_test), batch_size=64)
    # Evaluate the model and store the results
    loss, accuracy, precision, recall = model.evaluate(X_test, y_test)
    f1_score = 2 * (precision * recall) / (precision + recall)
    results.append({
        'Optimizer': optimizer.__class__.__name__,
        'Test Loss': loss,
        'Test Accuracy': accuracy,
        'Test Precision': precision,
        'Test Recall': recall,
        'Test F1 Score': f1_score
    })

# Create a DataFrame from the results
df_oversampled = pd.DataFrame(results)
```

```python
results = []  # List to store the results for each optimizer

# Iterate over each optimizer and train the model
for optimizer in optimizers:
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
    
    history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=64, class_weight=class_weights_dict)
    
    # Evaluate the model and store the results
    loss, accuracy, precision, recall = model.evaluate(X_test, y_test)
    f1_score = 2 * (precision * recall) / (precision + recall)
    results.append({
        'Optimizer': optimizer.__class__.__name__,
        'Test Loss': loss,
        'Test Accuracy': accuracy,
        'Test Precision': precision,
        'Test Recall': recall,
        'Test F1 Score': f1_score
    })

# Create a DataFrame from the results
df_class_weights = pd.DataFrame(results)
```
